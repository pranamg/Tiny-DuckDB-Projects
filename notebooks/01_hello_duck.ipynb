# Chapter 01: Hello Duck

Welcome to your first DuckDB project! This notebook will help you explore DuckDB basics.

```python
import duckdb

# Create an in-memory connection
conn = duckdb.connect()
```

## Basic SELECT

The simplest query - select a literal value:

```python
# Select a string literal
conn.execute("SELECT 'Hello, DuckDB!'").fetchone()
```

```python
# Select with an alias
conn.execute("SELECT 'Hello, DuckDB!' AS greeting").fetchone()
```

## Working with Numbers

```python
# Basic arithmetic
conn.execute("SELECT 2 + 2 AS sum, 10 * 5 AS product, 100 / 4 AS quotient").fetchone()
```

## Using DataFrames

DuckDB integrates nicely with pandas:

```python
# Get result as DataFrame
conn.execute("SELECT 'Hello' AS greeting, 42 AS answer").df()
```

```python
# Clean up
conn.close()
```
