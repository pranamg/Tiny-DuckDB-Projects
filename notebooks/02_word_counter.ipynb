# Chapter 02: Word Counter

Learn to count word frequencies using DuckDB's string functions and aggregations.

```python
import duckdb
conn = duckdb.connect()
```

## String Functions

DuckDB has powerful string manipulation functions:

```python
# Split a string into words
conn.execute("SELECT string_split('hello world foo bar', ' ')").fetchone()
```

```python
# Unnest to get individual words
conn.execute("""
    SELECT unnest(string_split('hello world foo bar', ' ')) AS word
""").fetchall()
```

## Counting with GROUP BY

```python
# Count word frequencies
text = "the quick brown fox jumps over the lazy dog the"

conn.execute(f"""
    WITH words AS (
        SELECT unnest(string_split(lower('{text}'), ' ')) AS word
    )
    SELECT word, COUNT(*) AS count
    FROM words
    WHERE word != ''
    GROUP BY word
    ORDER BY count DESC
""").df()
```
