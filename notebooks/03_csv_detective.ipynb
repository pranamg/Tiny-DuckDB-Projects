# Chapter 03: CSV Detective

Learn to read and query CSV files directly with DuckDB.

```python
import duckdb
conn = duckdb.connect()
```

## Reading CSV Files

DuckDB can read CSV files directly in queries:

```python
# Read the sample scores file
conn.execute("SELECT * FROM read_csv('../inputs/sample_scores.csv')").df()
```

## Filtering with WHERE

```python
# Find high scores
conn.execute("""
    SELECT * FROM read_csv('../inputs/sample_scores.csv')
    WHERE score > 1500
    ORDER BY score DESC
""").df()
```

```python
# Find a specific player
conn.execute("""
    SELECT * FROM read_csv('../inputs/sample_scores.csv')
    WHERE player = 'Alice'
""").df()
```

## Aggregations

```python
# Find highest score per game
conn.execute("""
    SELECT game, MAX(score) as highest, MIN(score) as lowest, AVG(score) as average
    FROM read_csv('../inputs/sample_scores.csv')
    GROUP BY game
""").df()
```
